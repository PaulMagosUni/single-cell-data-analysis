```python
import pandas as pd

df = pd.read_csv('results/aggregate/tabula-muris-heart/markers.csv')

num_clusters = 11

df.head()
```

```python
diz = {}

# for each row in df
for tool in df['tool'].unique():
    diz[tool] = {}
    tool_df = df[df['tool'] == tool]
    for cluster in df['cluster'].unique():
        cluster_df = tool_df[tool_df['cluster'] == cluster]
        diz[tool][cluster] = cluster_df['gene'].tolist()
        
diz
```

```python
for cluster in range(1, 12):
    val = len(set(diz['COTAN'][cluster] ).intersection(set(diz['scvi-tools'][cluster])))
    print('cluster', cluster, val, val/50 * 100, '%')
```
